```python
import numpy as np
data = np.array([[26,50000],
                 [29,70000],
                 [34,55000],
                 [31,41000],
                 [33,53000],
                 [28,49000],
                 [30,61000],
                 [35,56000]])
```

```python
#Normalization

from sklearn.preprocessing import MinMaxScaler

scaler  = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
scaled_data
```

```python
#Standization

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
scaled_data
```
