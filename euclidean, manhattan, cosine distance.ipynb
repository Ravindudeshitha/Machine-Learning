# Euclidean distance

```python
a=(5,3)
b= (2,4)
```

```python
from scipy.spatial import distance
```

```python
d = distance.euclidean(a,b)
print(d)
```

# Manhattan distance

```python
d2 = distance.cityblock(a,b)
print(d2)
```

# Cosine distance

```python
d3 = distance.cosine(a,b)
print(d3)
```
